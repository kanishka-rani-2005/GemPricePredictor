**Numerical and Categorical columns seperation**

**EDA (Exploratory Data Anaylsis)**

Boxplots

```python

for col in num_cols:
    plt.figure(figsize=(6,3))
    data=pd.DataFrame(df[col])
    data.boxplot()
    plt.tight_layout()
    plt.show()
```

Heatmap for numerical columns

```python
data=df[num_cols]
corr=data.corr()
sns.heatmap(corr,annot=True)
plt.tight_layout()
plt.show()
```
